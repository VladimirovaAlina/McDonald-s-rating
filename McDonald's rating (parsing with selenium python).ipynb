```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

import pandas as pd
```

```python
def infor(df):
    n = max(df['text_all'].keys()) if len(df['text_all'].keys()) != 0 else 0
    
    for blok in driver.find_elements_by_class_name('review'):

        for nm in blok.find_elements_by_class_name('consumer-information__name'):
            df['name'][n]=nm.text
        for tm in blok.find_elements_by_class_name('review-date--tooltip-target'):
            df['times'][n]=tm.text  
        for cntr in blok.find_elements_by_class_name('consumer-information__location'):
            df['country'][n]=cntr.text 
#         for rtng in blok.find_elements_by_class_name('star-rating.star-rating--medium'):
#             df['raiting'][n]=rtng.text
        for rtng in blok.find_elements_by_class_name('star-rating.star-rating--medium'):
            df['raiting'][n] = rtng.find_element_by_tag_name('img').get_attribute('alt')
       
        for lnk in blok.find_elements_by_class_name('link.link--large.link--dark'):
            df['link_name'][n]=lnk.text

        for txt in blok.find_elements_by_class_name('review-content__text'):
            try:
                df['text_all'][n]=txt.text    
            except TypeError:
                df['text_all'][n] = ''
        n+=1 
```

```python
driver = webdriver.Chrome(executable_path='./chromedriver')
driver.get("https://ru.trustpilot.com")
k = driver.find_elements_by_class_name('searchInputField___3e9zp')[0]

but = driver.find_elements_by_class_name('button___3ZeMh.large___3SFYn.primary___3u9Ht')[0]

k.send_keys('www.mcdonalds.com')
but.click()

# driver.find_elements_by_class_name('search-result-heading')[0].click()
driver.find_elements_by_class_name('button.button--primary.button--flexible.show-all-reviews')[0].click()
time.sleep(2)

df = {'name': {}, 'times': {}, 'country': {}, 'raiting': {}, 'link_name': {}, 'text_all': {}}

infor(df)
    
while True:
    driver.find_elements_by_class_name('button.button--primary.next-page')[0].click()
    time.sleep(2)
    infor(df)
```

```python
data = pd.DataFrame(df)
```
